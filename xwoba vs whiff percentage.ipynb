```python
scat = plt.scatter(dd["xwoba"], dd["swing_and_miss"],
                   c=dd.woba, cmap='seismic', s=75)
plt.title("xwOBA vs Swing & Miss% vs wOBA in 2018 Season", fontsize=15)
plt.xlabel("xwOBA", fontsize=13)
plt.ylabel("Swing and Miss Percentage", fontsize=13)
plt.xlim(0.220, 0.420)
plt.ylim(11.0, 38.0)

cl = plt.colorbar(scat)
cl.set_label("wOBA")

p = sns.regplot("xwoba", "swing_and_miss", data=dd, scatter=False)

p.figure.set_size_inches(18,8)

for line in range(0, dd.shape[0]):
    p.text(dd.xwoba[line]+0.00095, dd.swing_and_miss[line], 
           dd.player_name[line], horizontalalignment='left', 
            size='medium', color='black', weight='bold')
```
