```python
info = data.loc[:, ['launch_speed', 'launch_angle', 'events']]
```

```python
info
```

```python
info['events'] = info['events'].replace({'force_out':'out'}, regex = True)
info['events'] = info['events'].replace({'field_out':'out'}, regex = True)
info['events'] = info['events'].replace({'sac_fly':'out'}, regex = True)
info['events'] = info['events'].replace({'sac_bunt':'out'}, regex = True)
info['events'] = info['events'].replace({'frounded_into_double_play':'out'}, regex = True)
info['events'] = info['events'].replace({'field_error':'out'}, regex = True)
info['events'] = info['events'].replace({'double_play':'out'}, regex = True)
info['events'] = info['events'].replace({'fielders_choice':'out'}, regex = True)
info['events'] = info['events'].replace({'fielders_choice_out':'out'}, regex = True)
info['events'] = info['events'].replace({'sac_fly_double_play':'out'}, regex = True)
info['events'] = info['events'].replace({'grounded_into_out':'out'}, regex = True)
info['events'] = info['events'].replace({'out_out':'out'}, regex = True)
#info.events.renameinfo[(info.events == 'force_out') | (info.events == 'field_out') |
    #(info.events == 'sac_fly') | (info.events == 'sac_bunt') |
     #(info.events == 'frounded_into_double_play') | (info.events == 'field_error') |
    #(info.events == 'double_play') | (info.events == 'fielders_choice') |
    #(info.events == 'fielders_choice_out') | (info.events == 'sac_fly_double_play') |
     #(info.events == 'triple_play')]
```

```python
#In areas where 85-120mph and 18-45degrees collides are so called the barrel zone
sns.lmplot('launch_speed', 'launch_angle', data=info, hue='events', fit_reg=False, size=8, aspect=1.5)

#plt.axvspan(85, 120, color='red', alpha=0.3)
#plt.axhspan(18, 45, color='red', alpha=0.3)
plt.axhline(linewidth=3, color='gray')
plt.title('Exit Velocity vs Launch Angle')
plt.xlabel('Exit Velocity')
plt.ylabel('Launch Angle')
plt.show()
```
